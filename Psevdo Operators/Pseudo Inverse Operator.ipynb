## Відкриття файлу для створення і тестування формул псевдообернених матриць

```python
from PIL import Image
import numpy as np
im = Image.open('x3.bmp')
matrix_image_x = np.array(im)
```

## Алгоритм, заснований на означеннi Мура-Пенроуза

```python
from numpy import linalg as LA
def mura_penrouza(a):
    return LA.pinv(a)
inv  = mura_penrouza(matrix_image_x)
print(np.allclose(matrix_image_x, np.dot(matrix_image_x, np.dot(inv, matrix_image_x))))
print(np.allclose(inv, np.dot(inv, np.dot(matrix_image_x, inv))))
```

##  Алгоритм на основi формули Гревiлля

```python
import itertools
from scipy import linalg
def grevilla(x):
    return linalg.pinv(x)
inv = grevilla(matrix_image_x)
print(np.allclose(matrix_image_x, np.dot(matrix_image_x, np.dot(inv, matrix_image_x))))
print(np.allclose(inv, np.dot(inv, np.dot(matrix_image_x, inv))))
```
